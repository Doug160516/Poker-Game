```python
import random #for shuffle functions and random numbers

import collections #for counters

from collections import Counter

from IPython.display import clear_output #if i need to clear output

from itertools import combinations

import time

values={'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':11,'Queen':12,'King':13,'Ace':14}
#will be able to return a value for the ranks of items
reverse_values={v:k for k,v in values.items()}
#if i need to get a rank back from a value
suits=('Hearts','Diamonds','Spades','Clubs')
#to append to a card object
ranks=('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
#to append to a card object

#gives values to hand rankings and can also give a hand ranking for a value
hand_rankings={"Royal Flush":10,"Straight Flush":9,"Four of a Kind":8,"Full House":7,"Flush":6,"Straight":5,"Three of a Kind":4,"Two Pair":3,"Pair":2,"High Card":1}
reverse_hand_rankings={v:k for k,v in hand_rankings.items()}
```

```python
class Card:
    
    def __init__(self,suit,rank):
        self.suit=suit #from the suits list
        self.rank=rank #from the ranks list
        self.value=values[rank] #returns a value based on a cards rank
    
    def __str__(self):
        return self.rank+" of "+self.suit #for hand_rank and other print functions
```

```python
class Deck:
    
    def __init__(self):
        self.all_cards=[] #to hold card objects
        
        for suit in suits:
            for rank in ranks:
                created_card=Card(suit,rank) #creates cards of all 4 suits with all 13 ranks
                self.all_cards.append(created_card) #attaches all cards in a compiled list
                
    def shuffle(self):
        random.shuffle(self.all_cards) #shuffles the deck
        
    def deal_one(self):
        return self.all_cards.pop(0) #deals a single card from the deck
```

```python
def sort(list1): 
    """
    Sorts lists with card objects from low to high based on their value attribute
    
    args:
        list1: List object with card objects inside
    Returns:
        None 
    Raises:
        AttributeError: If list1 does not contain Card objects
    
    """
    list1.sort(key=lambda x:x.value)
```

```python
def print_kickers(player,num,y): #prints out a string to represent the appropriate amount of kickers for comparison func
    """
    Assigns a string to player.kicker_string to represent the appropriate amount of kickers for a comparison function
    
    args:
        player: Player object
        num: Int-number of kickers that need to be shown
        y:True or False-True for Four of a Kind, Three of a kind,two pair and Pair. False for Flush and High Card
    Returns:
        none
    Raises:
        AttributeError:If player is not a Player object type
        NameError:If y not equal to True or False
    
    """
    #if hand==four of a kind, three of a kind, two pair or pair then y should be true
    #for flushes and high cards, y should be false
    player.kicker_string=""#To clear kicker string
    list2=[] #empty list to append to 
    for items in player.hand_kickers: #takes kickers from hand
        list2.append(items.rank) #attaches the rank from the kickers list
    list2.reverse() #reverses the list so index[0] will be the highest value
    final_list=[]#third list to append to 
    if y==True: #if hand==four of a kind, three of a kind, two pair, or pair where kickers are irrelevant to hand type
        for items in list2[0:num:1]: #takes the list up to the appropriate index
            final_list.append(items)
    
        print_string=str(final_list) #makes a string of the list
        print_string=print_string.replace("[","") #replaces [] and ""
        print_string=print_string.replace("]","") #replaces [] and ""
        print_string=print_string.replace("'","") #replaces [] and ""
        print_string+=" Kicker"
        player.kicker_string=print_string #assigns kicker string to player object
    else: #for flushes and high cards where all five cards are needed to check value
        for items in list2[1:num+1:1]: #misses out the first kicker so the player.hand_string doesn't make it weird
            final_list.append(items) #attaches appropriate items
    
        print_string=str(final_list) #makes a string of the list
        print_string=print_string.replace("[","") #replaces [] and ""
        print_string=print_string.replace("]","") #replaces [] and ""
        print_string=print_string.replace("'","") #replaces [] and ""
        print_string+=" Kicker"
        player.kicker_string=print_string #assigns kicker string to player object
```

```python
def straight_flush_check(player): #checks players hand for straight flush
    """
    Checks player hand for a straight flush
    
    args:
        player:Player object type
    Returns:
        None
    Raises:
        AttributeError:If player is not a Player object type
        
    
    """
    flush_hand=[] #to attach player hand without accidentally altering it 
    flush_suit_list=[] #list of suits in player.card_hand
    for cards in player.card_hand:
        flush_suit_list.append(cards.suit) #appends suits to the flush_suit_list
    flush_suit_counter=Counter(flush_suit_list) #counts number of instances of each suit in player_card_hand
    for cards in player.card_hand:
        if flush_suit_counter[cards.suit]>=5: #if there are 5 or more of the same suit it is attached to flush_hand 
            flush_hand.append(cards)
        else:
            pass
    if len(flush_hand)>=5: #if flush_hand has 5 or more then a flush has occured
        checking=True #to create a break in case 6 or 7 cards still line up as a straight flush
        while checking==True: #creates break
            sort(flush_hand) #sorting flush_hand from low at [0] to high at [-1]
            try: #royal flush check
                if flush_hand[-1].value==14 and flush_hand[-2].value==13 and flush_hand[-3].value==12 and flush_hand[-4].value==11 and flush_hand[-5].value==10:
                    player.final_hand.append(flush_hand[-1])
                    player.final_hand.append(flush_hand[-2])
                    player.final_hand.append(flush_hand[-3])
                    player.final_hand.append(flush_hand[-4])
                    player.final_hand.append(flush_hand[-5])
                    
                    sort(player.final_hand) #probably not needed but just to be safe
                    
                    player.hand_value=10 #gives number value to hand
                    player.hand_string=f"{reverse_hand_rankings[player.hand_value]}" #prints out the name of the hand
                    checking=False
                    break
            except IndexError: #if the list isn't long enough it will skip past the error.
                pass
            else:
                pass
            try: #high end straight flush
                if flush_hand[-1].value==flush_hand[-2].value+1 and flush_hand[-2].value==flush_hand[-3].value+1 and flush_hand[-3].value==flush_hand[-4].value+1 and flush_hand[-4].value==flush_hand[-5].value+1:
                    player.final_hand.append(flush_hand[-1])
                    player.final_hand.append(flush_hand[-2])
                    player.final_hand.append(flush_hand[-3])
                    player.final_hand.append(flush_hand[-4])
                    player.final_hand.append(flush_hand[-5])
                    
                    sort(player.final_hand) #probably not needed
                        
                    player.hand_value=9 #gives value of straight flush
                    player.nested_hand_value=player.final_hand[-1].value #gives the value of the particular straight flush
                    player.hand_string=f"{reverse_hand_rankings[player.hand_value]}-{player.final_hand[-1].rank} High"
                    checking=False
                    break
            except IndexError: #passes if there is no appropriate index
                pass
            else:
                pass
            try: #mid range straight flush-same as above but with indexes changed
                if flush_hand[-2].value==flush_hand[-3].value+1 and flush_hand[-3].value==flush_hand[-4].value+1 and flush_hand[-4].value==flush_hand[-5].value+1 and flush_hand[-5].value==flush_hand[-6].value+1:
                    player.final_hand.append(flush_hand[-2])
                    player.final_hand.append(flush_hand[-3])
                    player.final_hand.append(flush_hand[-4])
                    player.final_hand.append(flush_hand[-5])
                    player.final_hand.append(flush_hand[-6])
                    
                    sort(player.final_hand)
                    
                    player.hand_value=9
                    player.nested_hand_value=player.final_hand[-1].value
                    player.hand_string=f"{reverse_hand_rankings[player.hand_value]}-{player.final_hand[-1].rank} High"
                    checking=False
                    break
            except IndexError:
                pass
            else:
                pass
            try: #low range straight flush-same as above but with indexes changed
                if flush_hand[-3].value==flush_hand[-4].value+1 and flush_hand[-4].value==flush_hand[-5].value+1 and flush_hand[-5].value==flush_hand[-6].value+1 and flush_hand[-6].value==flush_hand[-7].value+1:
                    player.final_hand.append(flush_hand[-3])
                    player.final_hand.append(flush_hand[-4])
                    player.final_hand.append(flush_hand[-5])
                    player.final_hand.append(flush_hand[-6])
                    player.final_hand.append(flush_hand[-7])
                    
                    sort(player.final_hand)
                    
                    player.hand_value=9
                    player.nested_hand_value=player.final_hand[-1].value
                    player.hand_string=f"{reverse_hand_rankings[player.hand_value]}-{player.final_hand[-1].rank} High"
                    checking=False
                    break
            except IndexError:
                pass
            else:
                pass
            try: #five high straight flush-same as royal flush but with indexes changed 
                if flush_hand[-1].value==14 and flush_hand[0].value==2 and flush_hand[1].value==3 and flush_hand[2].value==4 and flush_hand[3].value==5:
                    player.final_hand.append(flush_hand[-1])
                    player.final_hand.append(flush_hand[0])
                    player.final_hand.append(flush_hand[1])
                    player.final_hand.append(flush_hand[2])
                    player.final_hand.append(flush_hand[3])
                    
                    sort(player.final_hand)
                    
                    player.hand_value=9
                    player.nested_hand_value=player.final_hand[-2].value #has to be -2 to return the 5 value instead of the ace value
                    player.hand_string=f"{reverse_hand_rankings[player.hand_value]}-{player.final_hand[-2].rank} High"
                    checking=False
                    break
            except IndexError:
                pass
            else:
                checking=False #if no straight is found then it breaks out of the loop
                break
    else: #passes if no straight flush is found
        pass
```

```python
def four_kind_check(player): #checks for four of a kind
    """
    Checks player hand for four of a kind
    
    args:
        player: A Player object type
    Returns:
        None
    Raises:
        AttributeError:If player is not a Player object type
    
    """
    four_of_a_kind_hand=[]# to keep four of a kind in so that i can access the index without accidentally extracting a kicker
    four_of_a_kind_kickers=[] #to keep all none four kind cards so that they can be used for kickers
    four_kind_list=[] #to keep values of cards so they can be counted
    #find if there are 4 of one value
    for cards in player.card_hand:
        four_kind_list.append(cards.value) #appends card values to the list
    #counter for values
    four_kind_value_counter=Counter(four_kind_list) #counts card values
    #assign four of a kind and kickers to seperate hand
    for cards in player.card_hand:
        if four_kind_value_counter[cards.value]<4: #cards with 3 or less instances of there value will be assigned as kickers
            four_of_a_kind_kickers.append(cards)
        elif four_kind_value_counter[cards.value]==4: #cards with 4 instances will be put into seperate list
            four_of_a_kind_hand.append(cards)
        else:
            pass
    if len(four_of_a_kind_hand)>0: #if four kind hand isn't empty then the player has four of a kind
        for items in four_of_a_kind_hand:
            player.final_hand.append(items) #appends the four cards first
        sort(four_of_a_kind_kickers)#sorts kickers
        player.final_hand.append(four_of_a_kind_kickers[-1]) #attaches highest value kicker to final_hand
        player.hand_kickers.append(four_of_a_kind_kickers[-1]) #attaches appropriate kicker to player.hand_kickers.
        sort(player.final_hand)#just for neatness, wont affect indexing or anything
        player.hand_value=8 #gives player the value for 4 of a kind
        player.nested_hand_value=four_of_a_kind_hand[0].value #nested value is the value of the four cards e.g aces, sevens, fours etc
        player.hand_string=f"{reverse_hand_rankings[player.hand_value]}-{four_of_a_kind_hand[0].rank}s" #gives a printable instance of the hand.
    else: 
        pass
```

```python
def full_house_check(player): #cheks for full house
    """
    Checks player hand for full house
    
    args:
        player: A Player object type
    Returns:
        None
    Raises:
        AttributeError: If player is not a Player Object type
    
    """
    #sorted by rank, values added to test list, if value counter is 3 added to full house 3, if 2 added to full house pair, sorted then for loop using -1 value
    full_house_list=[]
    full_house_threes=[]
    full_house_pairs=[]
            
    for cards in player.card_hand: #appends values to list so they can be counted
        full_house_list.append(cards.value)
    #counter for values in player_card_hand
    full_house_value_counter=Counter(full_house_list)
        #for loop to add to threes and pairs list
    for cards in player.card_hand:
        if full_house_value_counter[cards.value]==3:
            full_house_threes.append(cards)
        elif full_house_value_counter[cards.value]==2:
            full_house_pairs.append(cards)
        else:
            pass
            #in case multiple 3 will append lower 3 to pair list
    if len(full_house_threes)>3:
        sort(full_house_threes)
        full_house_pairs.append(full_house_threes[0])
        full_house_pairs.append(full_house_threes[1])
    else:
        pass
                
        #if 3 list and pair list aren't empty then full house is true
    if len(full_house_threes)>0 and len(full_house_pairs)>0: 
        sort(full_house_threes)
        sort(full_house_pairs)
        player.final_hand.append(full_house_threes[-1])
        player.final_hand.append(full_house_threes[-2])
        player.final_hand.append(full_house_threes[-3])
        player.final_hand.append(full_house_pairs[-1])
        player.final_hand.append(full_house_pairs[-2])
        player.hand_kickers.append(full_house_pairs[-1])
        player.hand_value=7
        player.nested_hand_value=full_house_threes[-1].value
        player.hand_string=f"{reverse_hand_rankings[player.hand_value]}:{full_house_threes[-1].rank}s and {full_house_pairs[-1].rank}s"
    else:
        pass
```

```python
def flush_check(player):
    """
    Checks player hand for flush
    
    args:
        player:A Player object type
    Returns:
        None
    Raises:
        AttributeError:If player is not a Player object type
    
    """
    flush_hand=[] #to collate flush hand
    flush_suit_list=[] #to calculate the number of each suit in hand
    for cards in player.card_hand:
        flush_suit_list.append(cards.suit)
    flush_suit_counter=Counter(flush_suit_list)
    for cards in player.card_hand:
        if flush_suit_counter[cards.suit]>=5: #if there are five or more of a suit the card goes to flush hand
            flush_hand.append(cards)
        else:
            pass
    if len(flush_hand)>=5:
        sort(flush_hand) #popping extra cards from hand
        if len(flush_hand)==6:
            flush_hand.pop(0)
        elif len(flush_hand)==7:
            flush_hand.pop(0)
            flush_hand.pop(0)
        else:
            pass
        for cards in flush_hand: #putting flush into player.final_hand
            player.final_hand.append(cards)
            
        
        sort(player.final_hand)
        
        player.hand_value=6
        for cards in player.final_hand:
            player.hand_kickers.append(cards)#for comparison check kickers -1:-5
        sort(player.hand_kickers)
        player.hand_string=f"{player.final_hand[-1].suit} {reverse_hand_rankings[player.hand_value]}:{player.final_hand[-1].rank} High"
    else:
        pass
```

```python
def straight_check(player):
    """
    Checks player hand for a straight
    
    args:
        player:A Player object type
    Returns:
        None
    Raises:
        AttributeError:If player is not a Player object type
    
    """
    #code is same as for straight flush mostly
    if len(player.card_hand)>=5:
        straight_hand=[]
        for cards in player.card_hand:
            straight_hand.append(cards)
        straight_list=[]
        for cards in straight_hand:
            straight_list.append(cards.rank)
        straight_counter=Counter(straight_list)
        for cards in straight_hand:
            straight_counter=Counter(straight_list)
            if straight_counter[cards.rank]>1:
                straight_hand.remove(cards)
                straight_list.remove(cards.rank)
        sort(straight_hand)
            #high end straight, code the same as for straight flush
        has_hand=False
        while has_hand==False:
            try:
                if straight_hand[-1].value==straight_hand[-2].value+1 and straight_hand[-2].value==straight_hand[-3].value+1 and straight_hand[-3].value==straight_hand[-4].value+1 and straight_hand[-4].value==straight_hand[-5].value+1:
                    if len(straight_hand)==6:
                        straight_hand.pop(0)
                    elif len(straight_hand)==7:
                        straight_hand.pop(0)
                        straight_hand.pop(0)
                    else:
                        pass
                    for cards in straight_hand:
                        player.final_hand.append(cards)
                    sort(player.final_hand)
                    player.hand_value=5
                    player.nested_hand_value=player.final_hand[-1].value
                    player.hand_string=f"{reverse_hand_rankings[player.hand_value]}-{player.final_hand[-1].rank} High"
                    has_hand=True
                    break

            except IndexError:
                pass
            else:
                pass
                #mid range straight,
            try:
                if straight_hand[-2].value==straight_hand[-3].value+1 and straight_hand[-3].value==straight_hand[-4].value+1 and straight_hand[-4].value==straight_hand[-5].value+1 and straight_hand[-5].value==straight_hand[-6].value+1:
                    if len(straight_hand)==6:
                        straight_hand.pop(-1)
                    elif len(straight_hand)==7:
                        straight_hand.pop(-1)
                        straight_hand.pop(0)
                    else:
                        pass
                    for cards in straight_hand:
                        player.final_hand.append(cards)
                    sort(player.final_hand)
                    player.hand_value=5
                    player.nested_hand_value=player.final_hand[-1].value
                    player.hand_string=f"{reverse_hand_rankings[player.hand_value]}-{player.final_hand[-1].rank} High"
                    has_hand=True
                    break
            except IndexError:
                pass
            else:
                pass
            #low end straight
            try:
                if straight_hand[-3].value==straight_hand[-4].value+1 and straight_hand[-4].value==straight_hand[-5].value+1 and straight_hand[-5].value==straight_hand[-6].value+1 and straight_hand[-6].value==straight_hand[-7].value+1:
                    if len(straight_hand)==6:
                        straight_hand.pop(-1)
                    elif len(straight_hand)==7:
                        straight_hand.pop(-1)
                        straight_hand.pop(-1)
                    else:
                        pass
                    for cards in straight_hand:
                        player.final_hand.append(cards)
                    sort(player.final_hand)
                    player.hand_value=5
                    player.nested_hand_value=player.final_hand[-1].value
                    player.hand_string=f"{reverse_hand_rankings[player.hand_value]}-{player.final_hand[-1].rank} High"
                    has_hand=True
                    break
            except IndexError:
                pass
            else:
                pass
            try:
                if straight_hand[-1].value==14 and straight_hand[0].value==2 and straight_hand[1].value==3 and straight_hand[2].value==4 and straight_hand[3].value==5: 
                    #pop so it only displays appropriate cards
                    if len(straight_hand)==6:
                        straight_hand.pop(-2)
                    elif len(straight_hand)==7:
                        straight_hand.pop(-2)
                        straight_hand.pop(-2)
                    else:
                        pass
                    for cards in straight_hand:
                        player.final_hand.append(cards)
                    sort(player.final_hand)
                    player.hand_value=5
                    player.nested_hand_value=player.final_hand[-2].value
                    player.hand_string=f"{reverse_hand_rankings[player.hand_value]}-{player.final_hand[-2].rank} High"
                    has_hand=True
                    break
            except IndexError:
                pass
            else:
                has_hand=True
                break
    else:
        pass
```

```python
def three_kind_check(player):
    """
    Checks player hand for Three of a kind
    
    args:
        player:A Player object type
    Returns:
        None
    Raises:
        AttributeError:If player is not a Player object type
    
    """
    
    #list to hold all values
    three_kind_list=[]
    #list to find triple cards
    three_kind_threes=[]
    #list to hold potential kickers
    three_kind_kickers=[]
    #append to three kind hand
    for cards in player.card_hand:
        three_kind_list.append(cards.value)
    #counter for values
    three_kind_value_counter=Counter(three_kind_list)
    for cards in player.card_hand:
        if three_kind_value_counter[cards.value]==3:
            three_kind_threes.append(cards)
        elif three_kind_value_counter[cards.value]<3:
            three_kind_kickers.append(cards)
        else:
            pass
    #don't need to check for multiple threes as it should have been caught by full house check
    if len(three_kind_threes)>0:
        player.final_hand.append(three_kind_threes[0])
        player.final_hand.append(three_kind_threes[1])
        player.final_hand.append(three_kind_threes[2])
        
        sort(three_kind_kickers)
        
        player.final_hand.append(three_kind_kickers[-1])
        player.final_hand.append(three_kind_kickers[-2])
        
        player.hand_kickers.append(three_kind_kickers[-1])
        player.hand_kickers.append(three_kind_kickers[-2])
        
        player.hand_value=4
        player.nested_hand_value=three_kind_threes[-1].value#first call for comparison func then go to hand_kickers
        player.hand_string=f"{reverse_hand_rankings[player.hand_value]}:{three_kind_threes[-1].rank}s"
    else:
        pass
```

```python
def two_pair_check(player):
    """
    Checks player hand for two pair
    
    args:
        player:A Player object type
    Returns:
        None
    Raises:
        AttributeError:If player is not a player object type
    
    """
    #list to hold all values
    two_pair_list=[]
    #list to hold paired hands
    two_pair_pairs=[]
    #list to hold potential kickers
    two_pair_kickers=[]
        
    for cards in player.card_hand:
        two_pair_list.append(cards.value)
    #counter for values
    two_pair_value_counter=Counter(two_pair_list)
    for cards in player.card_hand:
        if two_pair_value_counter[cards.value]==2:
            two_pair_pairs.append(cards)
        elif two_pair_value_counter[cards.value]==1:
            two_pair_kickers.append(cards)
        else:
            pass
    #puts a potential third pair back in with kickers
    if len(two_pair_pairs)>=4:
        sort(two_pair_pairs)
        if len(two_pair_pairs)>4:
            two_pair_kickers.append(two_pair_pairs[0])
            two_pair_kickers.append(two_pair_pairs[1])
        else:
            pass
        sort(two_pair_kickers)
        player.final_hand.append(two_pair_pairs[-1])
        player.final_hand.append(two_pair_pairs[-2])
        player.final_hand.append(two_pair_pairs[-3])
        player.final_hand.append(two_pair_pairs[-4])
        player.final_hand.append(two_pair_kickers[-1])
        
        player.hand_kickers.append(two_pair_pairs[-3])#for comparison check kickers[0] for second pair
        player.hand_kickers.append(two_pair_kickers[-1])#for comparison check kickers[-1] for kicker
        
        player.hand_value=3
        player.nested_hand_value=two_pair_pairs[-1].value
        player.hand_string=f"{reverse_hand_rankings[player.hand_value]}:{two_pair_pairs[-1].rank}s and {two_pair_pairs[-3].rank}s"
    else:
        pass
```

```python
def pair_check(player):
    """
    Checks player hand for a pair
    
    args:
        player:A Player object type
    Returns:
        None
    Raises:
        AttributeError:If player is not a player object type
    
    """
    #list to hold all values
    pair_list=[]
    #list to hold pair
    pair_pair=[]
    #list to hold kickers
    pair_kickers=[]

        
    #append values to list
    for cards in player.card_hand:
        pair_list.append(cards.value)
    #counter for values
    pair_value_counter=Counter(pair_list)
    #append pairs and kickers to appropriate lists, else statement might have been sufficient but I'll use an elif for readability
    for cards in player.card_hand:
        if pair_value_counter[cards.value]==2:
            pair_pair.append(cards)
        elif pair_value_counter[cards.value]==1:
            pair_kickers.append(cards)
        else:
            pass
    if len(pair_pair)==2:
        if len(player.card_hand)>2:
            sort(pair_kickers)
            player.final_hand.append(pair_pair[0])
            player.final_hand.append(pair_pair[1])
            player.final_hand.append(pair_kickers[-1])
            player.final_hand.append(pair_kickers[-2])
            player.final_hand.append(pair_kickers[-3])
        
            player.hand_kickers.append(pair_kickers[-1])#for comparison check -1 - -3 for high to low
            player.hand_kickers.append(pair_kickers[-2])#for comparison check -1 - -3 for high to low
            player.hand_kickers.append(pair_kickers[-3])#for comparison check -1 - -3 for high to low
        
            sort(player.hand_kickers)
        else:
            player.final_hand.append(pair_pair[0])
            player.final_hand.append(pair_pair[1])
        
        player.hand_value=2
        player.nested_hand_value=pair_pair[0].value
        player.hand_string=f"{reverse_hand_rankings[player.hand_value]}:{pair_pair[-1].rank}s"
    else:
        pass
```

```python
def high_card_check(player):
    """
    Assigns high card to a player hand
    
    args:
        player:A Player object type
    Returns:
        None
    Raises:
        AttributeError:If player is not a player object type
    
    """
    sort(player.card_hand) #sorts player card low to high
    
    #Appends 5 highest ranked cards to final hand
    if len(player.card_hand)>2:
        player.final_hand.append(player.card_hand[-1])
        player.final_hand.append(player.card_hand[-2])
        player.final_hand.append(player.card_hand[-3])
        player.final_hand.append(player.card_hand[-4])
        player.final_hand.append(player.card_hand[-5])
    else:
        player.final_hand.append(player.card_hand[-1])
        player.final_hand.append(player.card_hand[-2])
    
    sort(player.final_hand) #sorts it low to high so that -1 will be the highest value
    
    for cards in player.final_hand:
        player.hand_kickers.append(cards)
        
    sort(player.hand_kickers)#-1 through -5 to check for parity.
    
    player.hand_value=1
    player.nested_hand_value=player.final_hand[-1].value
    player.hand_string=f"{reverse_hand_rankings[player.hand_value]}:{player.final_hand[-1].rank}"
```

```python
def hand_score(player):
    """
    Gives a integer value to a players hand
    
    args:
        player:A Player object type
    Returns:
        None
    Raises:
        AttributeError:If player is not a Player object type
    
    """
    if player.hand_value==1:#high card
        if len(player.final_hand)>2:
            player.hand_score=(14**4*player.hand_kickers[-1].value)+(14**3*player.hand_kickers[-2].value)+(14**2*player.hand_kickers[-3].value)+(14*player.hand_kickers[-4].value)+(player.hand_kickers[-5].value)
        else:
            player.hand_score=(14**4*player.hand_kickers[-1].value)+(14**3*player.hand_kickers[-2].value)
    elif player.hand_value==2:#pair
        if len(player.final_hand)>2:
            player.hand_score=1000000+(14**3*player.nested_hand_value)+(14**2*player.hand_kickers[-1].value)+(14*player.hand_kickers[-2].value)+(player.hand_kickers[-3].value)
        else:
            player.hand_score=1000000+(14**3*player.nested_hand_value)
    elif player.hand_value==3:#two pair
        player.hand_score=2000000+(14**2*player.nested_hand_value)+(14*player.hand_kickers[0].value)+(player.hand_kickers[1].value)
    elif player.hand_value==4:#three kind
        player.hand_score=3000000+(14**2*player.nested_hand_value)+(14*player.hand_kickers[-1].value)+(player.hand_kickers[-2].value)
    elif player.hand_value==5:#straight
        player.hand_score=4000000+((14**3*player.nested_hand_value)*10)
    elif player.hand_value==6:#flush
        player.hand_score=5000000+(14**4*player.hand_kickers[-1].value)+(14**3*player.hand_kickers[-2].value)+(14**2*player.hand_kickers[-3].value)+(14*player.hand_kickers[-4].value)+(player.hand_kickers[-5].value)
    elif player.hand_value==7:#full house
        player.hand_score=6000000+(14**2*player.nested_hand_value)+(14*player.hand_kickers[0].value)
    elif player.hand_value==8:#four of a kind
        player.hand_score=7000000+(14**2*player.nested_hand_value)+(14*player.hand_kickers[0].value)
    elif player.hand_value==9:#straight flush
        player.hand_score=8000000+((14**3*player.nested_hand_value)*10)
    elif player.hand_value==10:#royal flush
        player.hand_score=9000000
    else:
        pass
```

```python
def hand_value_check(player):
    """
    Finds the best 5 card combination from seven cards
    
    args:
        player:
            A Player object type
    Returns:
        None
    Raises:
        AttributeError:If player is not a player object type
        
    
    """
    #clears any player variables
    player.final_hand.clear()
    player.hand_value=0
    player.nested_hand_value=0
    player.hand_kickers.clear()
    player.hand_string="" 
    
    #maybe should have player.final_hand.clear()?
    #to break when a hand is found
    def end_check(player):
        """
        Returns a boolean value to break out of a loop
        """
        if len(player.final_hand)>0:
            return True
        else:
            return False
    #to keep checking until the highest hand is found
    has_hand=False
    while has_hand==False:
        straight_flush_check(player)
        if end_check(player)==True:
            has_hand=True
            break
        else:
            pass
        
        four_kind_check(player)
        if end_check(player)==True:
            has_hand=True
            break
        else:
            pass
        
        full_house_check(player)
        if end_check(player)==True:
            has_hand=True
            break
        else:
            pass
        
        flush_check(player)
        if end_check(player)==True:
            has_hand=True
            break
        else:
            pass
        
        straight_check(player)
        if end_check(player)==True:
            has_hand=True
            break
        else:
            pass
        
        three_kind_check(player)
        if end_check(player)==True:
            has_hand=True
            break
        else:
            pass
        
        two_pair_check(player)
        if end_check(player)==True:
            has_hand=True
            break
        else:
            pass
        
        pair_check(player)
        if end_check(player)==True:
            has_hand=True
            break
        else:
            pass
        
        high_card_check(player)
        if end_check(player)==True:
            has_hand=True
            break
        else:
            pass
    hand_score(player)
```

```python
def find_winner(player1,player2):
    """
    Find the best 5 card hand from two sets of seven cards and find the winner between the two returning 0-tie,1-player1 win or 2-player2 win
    
    args:
        player1:A Player object type
        player2:A Player object type
    Returns:
        0-Tied Hand
        1-player1 wins
        2-player2 wins
    Raises:
        AttributeError:If player1 or player2 are not Player object types
    
    """
    hand_value_check(player1)
    hand_value_check(player2)
    #checks player and oppo hand values against each other then checks all kicker or value levels
    winnerfound=False
    while winnerfound==False: #add breaks and winnnerfound=false
        if player1.hand_value>player2.hand_value: #player1 wins
            print(f"{player1.name} wins:{player1.hand_string} beats {player2.hand_string}") #player wins with better hand class
            return 1
            winnerfound=True
            break
        elif player1.hand_value<player2.hand_value: #player2 wins
            print(f"{player2.name} wins:{player2.hand_string} beats {player1.hand_string}") #opponent wins with better hand class
            return 2
            winnerfound=True
            break
        else:
            if player1.hand_value==10:#Royal flush
                print(f"Split Pot:{player1.name} and {player2.name} tied with {player1.hand_string}") #royal flush-tie
                return 0
                winnerfound=True
                break

            elif player1.hand_value==9:#Straight flush
                if player1.nested_hand_value>player2.nested_hand_value:#player1 wins straight flush
                    print(f"{player1.name} wins:{player1.hand_string} beats {player2.hand_string}")
                    return 1
                    winnerfound=True
                    break
                elif player1.nested_hand_value<player2.nested_hand_value: #player2 wins straight flush
                    print(f"{player2.name} wins:{player2.hand_string} beats {player1.hand_string}")
                    return 2
                    winnerfound=True
                    break
                else:
                    print(f"Split Pot:{player1.name} and {player2.name} tied with {player1.hand_string}") #straight flush-tie
                    return 0
                    winnerfound=True
                    break

            elif player1.hand_value==8:#Four of a kind
                if player1.nested_hand_value>player2.nested_hand_value:#player1 wins four of a kind
                    print(f"{player1.name} wins:{player1.hand_string} beats {player2.hand_string}") #four of a kind-player wins
                    return 1
                    winnerfound=True
                    break
                elif player1.nested_hand_value<player2.nested_hand_value:#player2 wins four of a kind
                    print(f"{player2.name} wins:{player2.hand_string} beats {player1.hand_string}") #four of a kind-opp wins
                    return 2
                    winnerfound=True
                    break
                else:
                    if player1.hand_kickers[0]>player2.hand_kickers[0]:#player1 wins four kind, kicker 1
                        print_kickers(player1,1,True)
                        print_kickers(player2,1,True)
                        print(f"{player1.name} wins:{player1.hand_string}-{player1.kicker_string} beats {player2.hand_string}-{player2.kicker_string}")
                        return 1
                        winnerfound=True
                        break
                    elif player1.hand_kickers[0]<player2.hand_kickers[0]:#player2 wins four kind, kicker 1
                        print_kickers(player1,1,True)
                        print_kickers(player2,1,True)
                        print(f"{player2.name} wins:{player2.hand_string}-{player2.kicker_string} beats {player1.hand_string}-{player2.kicker_string}")
                        return 2
                        winnerfound=True
                        break
                    else:
                        print(f"Split Pot:{player1.name} and {player2.name} tied with {player1.hand_string}") #four of a kind tie
                        return 0
                        winnerfound=True
                        break


            elif player1.hand_value==7:#Full House
                if player1.nested_hand_value>player2.nested_hand_value:#player 1 wins full house
                    print(f"{player1.name} wins:{player1.hand_string} beats {player2.hand_string}")
                    return 1
                    winnerfound=True
                    break
                elif player1.nested_hand_value<player2.nested_hand_value: #player 2 wins full house
                    print(f"{player2.name} wins:{player2.hand_string} beats {player1.hand_string}")
                    return 2
                    winnerfound=True
                    break
                else:
                    if player1.hand_kickers[0].value>player2.hand_kickers[0].value: #player 1 wins full house-kicker
                        print(f"{player1.name} wins:{player1.hand_string} beats {player2.hand_string}")
                        return 1
                        winnerfound=True
                        break
                    elif player1.hand_kickers[0].value<player2.hand_kickers[0].value: #player 2 wins full house-kicker
                        print(f"{player2.name} wins:{player2.hand_string} beats {player1.hand_string}")
                        return 2
                        winnerfound=True
                        break
                    else:#full house tie
                        print(f"Split Pot:{player1.name} and {player2.name} tied with {player1.hand_string}") #full house-tie
                        return 0
                        winnerfound=True
                        break

            elif player1.hand_value==6:#flush
                    if player1.hand_kickers[-1].value>player2.hand_kickers[-1].value:
                        print(f"{player1.name} wins:{player1.hand_string} beats {player2.hand_string}") #flush-player1 1 card win
                        return 1
                        winnerfound=True
                        break
                    elif player1.hand_kickers[-1].value<player2.hand_kickers[-1].value:
                        print(f"{player2.name} wins:{player2.hand_string} beats {player1.hand_string}") #flush-player2 1 card win
                        return 2
                        winnerfound=True
                        break
                    else:
                        if player1.hand_kickers[-2].value>player2.hand_kickers[-2].value:
                            print_kickers(player1,1,False)
                            print_kickers(player2,1,False)
                            print(f"{player1.name} wins:{player1.hand_string}-{player1.kicker_string} beats {player2.hand_string}-{player2.kicker_string}")#flush-player1 2 card win
                            return 1
                            winnerfound=True
                            break
                        elif player1.hand_kickers[-2].value<player2.hand_kickers[-2].value:
                            print_kickers(player1,1,False)
                            print_kickers(player2,1,False)
                            print(f"{player2.name} wins:{player2.hand_string}-{player2.kicker_string} beats {player1.hand_string}-{player1.kicker_string}")#flush-player2 2 card win
                            return 2
                            winnerfound=True
                            break
                        else:
                            if player1.hand_kickers[-3].value>player2.hand_kickers[-3].value:
                                print_kickers(player1,2,False)
                                print_kickers(player2,2,False)
                                print(f"{player1.name} wins:{player1.hand_string}-{player1.kicker_string} beats {player2.hand_string}-{player2.kicker_string}")#flush-player1 3 card win
                                return 1
                                winnerfound=True
                                break
                            elif player1.hand_kickers[-3].value<player2.hand_kickers[-3].value:
                                print_kickers(player1,2,False)
                                print_kickers(player2,2,False)
                                print(f"{player2.name} wins:{player2.hand_string}-{player2.kicker_string} beats {player1.hand_string}-{player1.kicker_string}")# flush-player2 3 card win
                                return 2
                                winnerfound=True
                                break
                            else:
                                if player1.hand_kickers[-4].value>player2.hand_kickers[-4].value:
                                    print_kickers(player1,3,False)
                                    print_kickers(player2,3,False)
                                    print(f"{player1.name} wins:{player1.hand_string}-{player1.kicker_string} beats {player2.hand_string}-{player2.kicker_string}")#flush-player1 4 card win
                                    return 1
                                    winnerfound=True
                                    break
                                elif player1.hand_kickers[-4].value<player2.hand_kickers[-4].value:
                                    print_kickers(player1,3,False)
                                    print_kickers(player2,3,False)
                                    print(f"{player2.name} wins:{player2.hand_string}-{player2.kicker_string} beats {player1.hand_string}-{player1.kicker_string}") #flush-player2 4 card win
                                    return 2
                                    winnerfound=True
                                    break
                                else:
                                    if player1.hand_kickers[-5].value>player2.hand_kickers[-5].value:
                                        print_kickers(player1,4,False)
                                        print_kickers(player2,4,False)
                                        print(f"{player1.name} wins:{player1.hand_string}-{player1.kicker_string} beats {player2.hand_string}-{player2.kicker_string}") #flush-player1 5 card win
                                        return 1
                                        winnerfound=True
                                        break
                                    elif player1.hand_kickers[-5].value<player2.hand_kickers[-5].value:
                                        print_kickers(player1,4,False)
                                        print_kickers(player2,4,False)
                                        print(f"{player2.name} wins:{player2.hand_string}-{player2.kicker_string} beats {player1.hand_string}-{player1.kicker_string}") #flush-player2 5 card win
                                        return 2
                                        winnerfound=True
                                        break
                                    else:
                                        print_kickers(player1,4,False)
                                        print_kickers(player2,4,False)
                                        print(f"Split pot:{player1.name} and {player2.name} tied with {player1.hand_string}-{player1.kicker_string}") #flush-tie
                                        return 0
                                        winnerfound=True
                                        break
           
            elif player1.hand_value==5:#straight
                if player1.nested_hand_value>player2.nested_hand_value:
                    print(f"{player1.name} wins:{player1.hand_string} beats {player2.hand_string}") #straight-player1 win
                    return 1
                    winnerfound=True
                    break
                elif player1.nested_hand_value<player2.nested_hand_value:
                    print(f"{player2.name} wins:{player2.hand_string} beats {player1.hand_string}") #straight-player2 win
                    return 2
                    winnerfound=True
                    break
                else:
                    print(f"Split pot:{player1.name} and {player2.name} tied with {player1.hand_string}") #straight-tie
                    return 0
                    winnerfound=True
                    break
                    
            elif player1.hand_value==4:#three of a kind
                if player1.nested_hand_value>player2.nested_hand_value:
                    print(f"{player1.name} wins:{player1.hand_string} beats {player2.hand_string}") #3 kind-player1 win
                    return 1
                    winnerfound=True
                    break
                elif player1.nested_hand_value<player2.nested_hand_value:
                    print(f"{player2.name} wins:{player2.hand_string} beats {player1.hand_string}") #3 kind-player2 win
                    return 2
                    winnerfound=True
                    break
                else:
                    if player1.hand_kickers[-1].value>player2.hand_kickers[-1].value:
                        print_kickers(player1,1,True)
                        print_kickers(player2,1,True)
                        print(f"{player1.name} wins:{player1.hand_string}-{player1.kicker_string} beats {player2.hand_string}-{player2.kicker_string}") #3 kind-player1 1 kicker win
                        return 1
                        winnerfound=True
                        break
                    elif player1.hand_kickers[-1].value<player2.hand_kickers[-1].value:
                        print_kickers(player1,1,True)
                        print_kickers(player2,1,True)
                        print(f"{player2.name} wins:{player2.hand_string}-{player2.kicker_string} beats {player1.hand_string}-{player1.kicker_string}") #3 kind-player2 1 kicker win
                        return 2
                        winnerfound=True
                        break
                    else:
                        if player1.hand_kickers[-2].value>player2.hand_kickers[-2].value:
                            print_kickers(player1,2,True)
                            print_kickers(player2,2,True)
                            print(f"{player1.name} wins:{player1.hand_string}-{player1.kicker_string} beats {player2.hand_string}-{player2.kicker_string}") #3 kind-player1 2 kicker win
                            return 1
                            winnerfound=True
                            break
                        elif player1.hand_kickers[-2].value<player2.hand_kickers[-2].value:
                            print_kickers(player1,2,True)
                            print_kickers(player2,2,True)
                            print(f"{player2.name} wins:{player2.hand_string}-{player2.kicker_string} beats {player1.hand_string}-{player1.kicker_string}") #3 kind-player2 2 kicker win
                            return 2
                            winnerfound=True
                            break
                        else:
                            print_kickers(player1,2,True)
                            print_kickers(player2,2,True)
                            print(f"Split pot:{player1.name} and {player2.name} tied with {player1.hand_string}-{player1.kicker_string}") #3 kind-tie
                            return 0
                            winnerfound=True
                            break
            
            elif player1.hand_value==3:#two pair
                if player1.nested_hand_value>player2.nested_hand_value:
                    print(f"{player1.name} wins:{player1.hand_string} beats {player2.hand_string}") #2 pair-top pair-player1 win
                    return 1
                    winnerfound=True
                    break
                elif player1.nested_hand_value<player2.nested_hand_value:
                    print(f"{player2.name} wins:{player2.hand_string} beats {player1.hand_string}") #2 pair-top pair-player2 win
                    return 2
                    winnerfound=True
                    break
                else:
                    if player1.hand_kickers[0].value>player2.hand_kickers[0].value:
                        print(f"{player1.name} wins:{player1.hand_string} beats {player2.hand_string}") #2 pair-second pair-player1 win
                        return 1
                        winnerfound=True
                        break
                    elif player1.hand_kickers[0].value<player2.hand_kickers[0].value:
                        print(f"{player2.name} wins:{player2.hand_string} beats {player1.hand_string}") #2 pair-second pair-player2 win
                        return 2
                        winnerfound=True
                        break
                    else:
                        if player1.hand_kickers[1].value>player2.hand_kickers[1].value:
                            print_kickers(player1,1,True)
                            print_kickers(player2,1,True) #need to test this
                            print(f"{player1.name} wins:{player1.hand_string}-{player1.kicker_string} beats {player2.hand_string}-{player2.kicker_string}") #2 pair-kicker-player win
                            return 1
                            winnerfound=True
                            break
                        elif player1.hand_kickers[1].value<player2.hand_kickers[1].value:
                            print_kickers(player1,1,True)
                            print_kickers(player2,1,True)
                            print(f"{player2.name} wins:{player2.hand_string}-{player2.kicker_string} beats {player1.hand_string}-{player1.kicker_string}") #2 pair-kicker-player2 win
                            return 2
                            winnerfound=True
                            break
                        else:
                            print(f"Split pot:{player1.name} and {player2.name} tied with {player1.hand_string}-{player1.kicker_string}") #2 pair-tie
                            return 0
                            winnerfound=True
                            break
            elif player1.hand_value==2:#pair
                if player1.nested_hand_value>player2.nested_hand_value:
                    print(f"{player1.name} wins:{player1.hand_string} beats {player2.hand_string}") #pair-player1 win
                    return 1
                    winnerfound=True
                    break
                elif player1.nested_hand_value<player2.nested_hand_value:
                    print(f"{player2.name} wins:{player2.hand_string} beats {player1.hand_string}") #pair-player2 win
                    return 2
                    winnerfound=True
                    break
                else:
                    if player1.hand_kickers[-1].value>player2.hand_kickers[-1].value:
                        print_kickers(player1,1,True)
                        print_kickers(player2,1,True)
                        print(f"{player1.name} wins:{player1.hand_string}-{player1.kicker_string} beats {player2.hand_string}-{player2.kicker_string}") #pair-first kicker-player1 win
                        return 1
                        winnerfound=True
                        break
                    elif player1.hand_kickers[-1].value<player2.hand_kickers[-1].value:
                        print_kickers(player1,1,True)
                        print_kickers(player2,1,True)
                        print(f"{player2.name} wins:{player2.hand_string}-{player2.kicker_string} beats {player1.hand_string}-{player1.kicker_string}") #pair-first kicker-player2 win
                        return 2
                        winnerfound=True
                        break
                    else:
                        if player1.hand_kickers[-2].value>player2.hand_kickers[-2].value:
                            print_kickers(player1,2,True)
                            print_kickers(player2,2,True)
                            print(f"{player1.name} wins:{player1.hand_string}-{player1.kicker_string} beats {player2.hand_string}-{player2.kicker_string}") #pair-second kicker-player1 win
                            return 1
                            winnerfound=True
                            break
                        elif player1.hand_kickers[-2].value<player2.hand_kickers[-2].value:
                            print_kickers(player1,2,True)
                            print_kickers(player2,2,True)
                            print(f"{player2.name} wins:{player2.hand_string}-{player2.kicker_string} beats {player1.hand_string}-{player1.kicker_string}") #pair-second kicker-player2 win
                            return 2
                            winnerfound=True
                            break
                        else:
                            if player1.hand_kickers[-3].value>player2.hand_kickers[-3].value:
                                print_kickers(player1,3,True)
                                print_kickers(player2,3,True)
                                print(f"{player1.name} wins:{player1.hand_string}-{player1.kicker_string} beats {player2.hand_string}-{player2.kicker_string}") #pair-third kicker-player1 win
                                return 1
                                winnerfound=True
                                break
                            elif player1.hand_kickers[-3].value<player2.hand_kickers[-3].value:
                                print_kickers(player1,3,True)
                                print_kickers(player2,3,True)
                                print(f"{player2.name} wins:{player2.hand_string}-{player2.kicker_string} beats {player1.hand_string}-{player1.kicker_string}") #pair-third kicker-player2 win
                                return 2
                                winnerfound=True
                                break
                            else:
                                print_kickers(player1,3,True)
                                print_kickers(player2,3,True)
                                print(f"Split pot:{player1.name} and {player2.name} tied with {player1.hand_string}-{player1.kicker_string}") #pair-tie
                                return 0
                                winnerfound=True
                                break

            else:#high card
                if player1.hand_kickers[-1].value>player2.hand_kickers[-1].value:
                    print(f"{player1.name} wins:{player1.hand_string} beats {player2.hand_string}") #high card-player1 win
                    return 1
                    winnerfound=True
                    break
                elif player1.hand_kickers[-1].value<player2.hand_kickers[-1].value:
                    print(f"{player2.name} wins:{player2.hand_string} beats {player1.hand_string}") #high card-player2 win
                    return 2
                    winnerfound=True
                    break
                else:
                    if player1.hand_kickers[-2].value>player2.hand_kickers[-2].value:
                        print_kickers(player1,1,False)
                        print_kickers(player2,1,False)
                        print(f"{player1.name} wins:{player1.hand_string}-{player1.kicker_string} beats {player2.hand_string}-{player2.kicker_string}") #high card-first kicker-player1 win
                        return 1
                        winnerfound=True
                        break
                    elif player1.hand_kickers[-2].value<player2.hand_kickers[-2].value:
                        print_kickers(player1,1,False)
                        print_kickers(player2,1,False)
                        print(f"{player2.name} wins:{player2.hand_string}-{player2.kicker_string} beats {player1.hand_string}-{player1.kicker_string}") #high card-first kicker-player2 win
                        return 2
                        winnerfound=True
                        break
                    else:
                        if player1.hand_kickers[-3].value>player2.hand_kickers[-3].value:
                            print_kickers(player1,2,False)
                            print_kickers(player2,2,False)
                            print(f"{player1.name} wins:{player1.hand_string}-{player1.kicker_string} beats {player2.hand_string}-{player2.kicker_string}") #high card-second kicker-player1 win
                            return 1
                            winnerfound=True
                            break
                        elif player1.hand_kickers[-3].value<player2.hand_kickers[-3].value:
                            print_kickers(player1,2,False)
                            print_kickers(player2,2,False)
                            print(f"{player2.name} wins:{player2.hand_string}-{player2.kicker_string} beats {player1.hand_string}-{player2.kicker_string}") #high card-second kicker-player2 win
                            return 2
                            winnerfound=True
                            break
                        else:
                            if player1.hand_kickers[-4].value>player2.hand_kickers[-4].value:
                                print_kickers(player1,3,False)
                                print_kickers(player2,3,False)
                                print(f"{player1.name} wins:{player1.hand_string}-{player1.kicker_string} beats {player2.hand_string}-{player2.kicker_string}") #high card-third kicker-player1 win
                                return 1
                                winnerfound=True
                                break
                            elif player1.hand_kickers[-4].value<player2.hand_kickers[-4].value:
                                print_kickers(player1,3,False)
                                print_kickers(player2,3,False)
                                print(f"{player2.name} wins:{player2.hand_string}-{player2.kicker_string} beats {player1.hand_string}-{player1.kicker_string}") #high card-third kicker-player2 win
                                return 2
                                winnerfound=True
                                break
                            else:
                                if player1.hand_kickers[-5].value>player2.hand_kickers[-5].value:
                                    print_kickers(player1,4,False)
                                    print_kickers(player2,4,False)
                                    print(f"{player1.name} wins:{player1.hand_string}-{player1.kicker_string} beats {player2.hand_string}-{player2.kicker_string}") #high card-fourth kicker-player1 win
                                    return 1
                                    winnerfound=True
                                    break
                                elif player1.hand_kickers[-5].value<player2.hand_kickers[-5].value:
                                    print_kickers(player1,4,False)
                                    print_kickers(player2,4,False)
                                    print(f"{player2.name} wins:{player2.hand_string}-{player2.kicker_string} beats {player1.hand_string}-{player1.kicker_string}") #high card-fourth kicker-player2 win
                                    return 2
                                    winnerfound=True
                                    break
                                else:
                                    print_kickers(player1,4,False)
                                    print_kickers(player2,4,False)
                                    print(f"Split pot:{player1.name} and {player2.name} tied with {player1.hand_string}-{player1.kicker_string}") #high card-tie
                                    return 0
                                    winnerfound=True
                                    break   
```
